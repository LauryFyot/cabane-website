```python
import os
from PIL import Image

# Dossiers à parcourir
folders = ["../assets/img", "../assets/img/activities"]

# Extensions d'image supportées
input_extensions = [".webp", ".webp", ".webp"]

# Options de qualité WebP (sans perte)
webp_lossless_options = {
    "format": "WEBP",
    "lossless": True,
    "quality": 100,
    "method": 6  # compression plus lente mais meilleure
}

def convert_to_webp(image_path):
    print(f'Converting {image_path}')
    base, ext = os.path.splitext(image_path)
    if ext.lower() not in input_extensions:
        return

    output_path = base + ".webp"
    if os.path.exists(output_path):
        print(f"Déjà converti : {output_path}")
        return

    try:
        with Image.open(image_path) as img:
            img.save(output_path, **webp_lossless_options)
            print(f"Converti : {image_path} → {output_path}")
    except Exception as e:
        print(f"Erreur avec {image_path} : {e}")

# Boucle sur les dossiers
for folder in folders:
    for root, _, files in os.walk(folder):
        print("test")
        for file in files:
            filepath = os.path.join(root, file)
            convert_to_webp(filepath)
```
